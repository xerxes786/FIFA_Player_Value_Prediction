```python
def position_simplifier(val):

    if val == 'RF' or val == 'ST' or val == 'LF' or val == 'RS' or val == 'LS' or val == 'CF' or val == 'RW' or val == 'LW':
        val = 'F'
        return val

    elif val == 'RCM' or val == 'LCM' or val == 'LDM' or val == 'CAM' or val == 'CDM' or val == 'RM' or val == 'LAM' or val == 'LM' or val == 'RDM' or val == 'CM' or val == 'RAM':
        val = 'M'
        return val


    elif val == 'RCB' or val == 'CB' or val == 'LCB' or val == 'LB' or val == 'RB' or val == 'RWB' or val == 'LWB':
        val = 'D'
        return val

    else:
        return val

df['Position'] = df['Position'].apply(position_simplifier)
```

```python
def ht_convert(val):
    foot = val.split("'")[0]
    inch = val.split("'")[1]
    h = (int(foot) * 30.48) + (int(inch)*2.54)
    return h

df['Height'] = df['Height'].apply(ht_convert)
```

```python
def join_nations(x):
    if x >= 68:
        return 'Class A'
    elif x >= 65 and x < 68:
        return 'Class B'
    elif x >= 20and x < 65:
        return 'Class C'
```

```python
countries_class = {}
for nation in df['Nationality'].unique():
    a = np.average(df[df['Nationality'] == nation]['Overall'])
    countries_class[nation] = join_nations(a)
```

```python
df['Nationality'].replace(countries_class, inplace=True)
```

```python
def join_clubs(x):
    if x >= 68:
        return 'Group A'
    elif x >= 64 and x < 68:
        return 'Group B'
    elif x >= 20 and x < 64:
        return 'Group C'
```

```python
clubs_group = {}
for club in df['Club'].unique():
    a = np.average(df[df['Club'] == club]['Overall'])
    clubs_group[club] = join_clubs(a)
```

```python
df['Club'].replace(clubs_group, inplace=True)
```
